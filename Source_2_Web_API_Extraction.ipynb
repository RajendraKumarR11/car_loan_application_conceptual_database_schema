```python
# Source_2_Web_API_Extraction
# This class containes the code for extracting popularity information related to cars through Twitter API.
# Detailed description availble through inline comments.

import twitter # pip install twitter
import pandas as pd
from urllib.parse import unquote
import pandas as pd


# Storing relevant keys and token information in desired variables
CONSUMER_KEY = '8kN7ITYYYK6bNkHKoViRPb7gJ'
CONSUMER_SECRET = 'g8IGSzKWckrYRzYjTgm8J2oNIywsmzcS8I9zW6WTrtIx28RGrE'
OAUTH_TOKEN = '1088871960698908674-WsDWHHSB2jcaH9zMwaW5mNgpwIzpGM'
OAUTH_TOKEN_SECRET = '0pHfCkkkjEhWeQbm6tZ9eWHRLXzIh4dyNimAHXmuvYg6j'
auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET,
                           CONSUMER_KEY, CONSUMER_SECRET)


# Initialising twitter api
twitter_api = twitter.Twitter(auth=auth)


# Reading the data extracted through web scraping from car website and stored in Source_Web_Scraping_Car_Specific_Data.csv
car_webscrap_input = pd.read_csv('Source_Web_Scraping_Car_Specific_Data.csv',encoding='latin-1');
 

# Extracting all values under column 'Car_Fullname' and storing it in car_name_list
car_name_list = list(car_webscrap_input['Car_Fullname'])



# Initialising four empty lists for storing car popularity related information
statusesCount = []
retweetsCount = []
favoritesCount = []
used_hashtags = []


#Iterating through list of all car names
for carname in car_name_list:
    hashtagname =''
    statuses = []
    hashtags = []
    result_hashtags = []
    
    # Creating a string resembling a probaable hashtag, which could have been made for tweeting about the respective car
    hashtags.append('#'+''.join(carname.split()))
                    
    # Creating a second tyoe of string resembling 
    # another probaable hashtag, which could have been made for tweeting about the respective car
    for item in carname.split():
        hashtagnamesub = '#'+item
        hashtagname = hashtagname + hashtagnamesub
    
    # Storing the hashtag information in hahtags list
    hashtags.append(hashtagname.strip())
    
    
    # Iterating through the hashtags list
    for tag in hashtags:
        # Setting the value of n as 5, so as to restrict the number of tweets, as more number of tweet retreival is causing
        #twitter to turn down the API request
        n = 5
        
        # Making the call to API and storing the results in search_results
        search_results = twitter_api.search.tweets(q=tag, count=n)
        
        #Picking up only the values under 'Statuses' column
        statuses = search_results['statuses']
        result_hashtags.append(tag)

        
    used_hashtags.append(",".join(result_hashtags))
    
    # Initializing the 'TotalFavoriteCount' and 'TotalRetweetCount' as zero
    TotalFavoriteCount = 0
    TotalRetweetCount = 0

    # Iterating number of timess equal to the number of statuses retreived
    for i in range(len(statuses)):
        
        # Filtering only English Language tweets
        if search_results['statuses'][i][u'lang']==u'en':
            
            # Storing the total favourite count and retweet count for all retrieved statuses
            TotalFavoriteCount = TotalFavoriteCount + statuses[i]['favorite_count']
            TotalRetweetCount =  TotalRetweetCount + statuses[i]['retweet_count']
            
    # Storing the calculated popularity parameters in relevant lists        
    statusesCount.append(len(statuses))
    retweetsCount.append(retweetsCount)
    favoritesCount.append(TotalFavoriteCount)
    
# Creating a dataframe to store all the popularity related information    
car_api_dataframe = pd.DataFrame({"Car_Fullname":car_name_list,
                        "Total_Status_Count":statusesCount,
                        "Total_Favorite_Count":TotalFavoriteCount,
                        "Total_Retweet_Count":TotalRetweetCount
                       })


# Exporting information from dataframe to a csv file  
car_api_dataframe.to_csv("Source_Web_API_Car_Popularity_Data.csv", index=False)


""""
CONCLUSIONS:

This code uses the list of car names, extracted through webscraing as input, tries to find the popularity parameters related 
to the same through Twitter API, stores the results in dataframe and exports it to a CSV file.

CONTRIBUTIONS:

RAJENDRA KUMAR RAJKUMAR had wriiten the code for the same and was reviewed by MONISH  HIRISAVE RAGHU

CITATIONS:

1. https://www.geeksforgeeks.org
2. https://github.com/nikbearbrown/INFO_6210

LICENSE:

Copyright <2019> <RAJENDRA KUMAR RAJKUMAR, MONISH  HIRISAVE RAGHU>

Permission is hereby granted, free of charge, to any person obtaining a copy of this software and associated documentation files (the "Software"), to deal in the Software without restriction, including without limitation the rights to use, copy, modify, merge, publish, distribute, sublicense, and/or sell copies of the Software, and to permit persons to whom the Software is furnished to do so, subject to the following conditions:

The above copyright notice and this permission notice shall be included in all copies or substantial portions of the Software.

THE SOFTWARE IS PROVIDED "AS IS", WITHOUT WARRANTY OF ANY KIND, EXPRESS OR IMPLIED, INCLUDING BUT NOT LIMITED TO THE WARRANTIES OF MERCHANTABILITY, FITNESS FOR A PARTICULAR PURPOSE AND NONINFRINGEMENT. IN NO EVENT SHALL THE AUTHORS OR COPYRIGHT HOLDERS BE LIABLE FOR ANY CLAIM, DAMAGES OR OTHER LIABILITY, WHETHER IN AN ACTION OF CONTRACT, TORT OR OTHERWISE, ARISING FROM, OUT OF OR IN CONNECTION WITH THE SOFTWARE OR THE USE OR OTHER DEALINGS IN THE SOFTWARE.


"""
```
